dropna()

```python
df=pd.read_csv('dummy.csv', sep=';')
```

```python
df.drop(columns='e', inplace=True)
df=df.head(10)
df
```

```python
df.dropna   # Drop every rows that contains a 'NaN'
```

```python
df.dropna(axis=0) # Removes all the row that contain a 'NaN'
```

```python
df.dropna(axis=1)  # Remove alll the  column that contain a 'NaN'
```

```python
df.dropna(axis=0, how='all')  # how='any' is the default [If any of the row having all are NaN will remove]
```

```python
df.dropna(axis=1, how='all')  # [If any of the column having all are NaN will remove]
```

fillna()

```python
data={'a': [1,2,3,None],
      'b': [5,None,7,8],
      'c': [9, None,11,None],
      'd': [13,14,None, 16]}

df=pd.DataFrame(data)
df
```

'ffill' & 'pad' uses the previous valid value to fill the next 'NaN'

'backfill' & 'bfill' use the last valid value to fill the previous 'NaN

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')  # If the last value is NaN. then it won't change
```

```python
df.fillna(method='ffill', axis=1)   # You can also define an axis
```

```python
df.fillna(method='bfill', axis=1) 
```

```python
df.fillna(value='???', limit=1, axis=1)   # Will replace only 1 value per row or column    
```

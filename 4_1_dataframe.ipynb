What are Dataframes?
. Dataframe is a 2-dimentional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table or a dict of series objects. It is generally the most commonly used pandas object.
. Essentially is just comprised of 2 or more series
. axis=0 refers to the index
. axis=1 refers to the columns

```python
import pandas as pd
```

```python
data={'a':[100,200,300], 'b':[400,500,600],'c':[700,800,900]}
df=pd.DataFrame(data)
df
```

```python
df['a']
```

```python
df=pd.DataFrame(data=[[100,200,300],[400,500,600],[700,800,900]],
                  index=['first', 'second','third'],
                  columns=['a','b','c'])
df
```

```python
df.a  # we can access by this also but not preferrred
```

2-Dataframe Methods

```python
series=pd.Series([100,200,300])
series
```

```python
data={'a':[100,200,300], 'b':[400,500,600],'c':[700,800,900]}
df=pd.DataFrame(data)
df
```

```python
series.sum()   # Sums each value and return a result
```

```python
df.sum()     # Sums each value per column and return a series
```

```python
df.sum(axis=1) # Will sum things up horizontaly (columns)
```

```python
df.sum(axis='index')  # Same as axis=0
```

```python
df.sum(axis='columns')   # Same as axis=1
```

```python
df.max(axis='columns')         
```

descrive() & info()

```python
cols=['Series_Title','IMDB_Rating','No_of_Votes','Gross','Released_Year','Genre']
```

```python
movies=pd.read_csv('Resources\imdb_top_1000.csv', usecols=cols)
movies
```

```python
movies.info()
```

```python
movies.describe().round()   # But what happened to Gross ?
```

```python
# Replace each comma with nothing
movies['Gross'] = movies['Gross'].str.replace(',','')
movies
```

```python
# Convert data into an integer so we can use it as numeric data
movies['Gross'].fillna(0, inplace=True)
movies['Gross']= movies['Gross'].astype('int64')
movies
```

```python
movies.describe().round() 
```

```python
movies.count()
```

```python
movies.mean(numeric_only=True).round()
```

```python
movies.max(numeric_only=True).round()
```

```python
movies.min(numeric_only=True).round()
```

```python
movies.describe(include=['O'])   # Includes every datatype if you specify 'all'
# Also provides 4 other fields
```

nlargest() & nsmallest()

```python
cols=['Series_Title','IMDB_Rating','No_of_Votes']
```

```python
movies=pd.read_csv('Resources\imdb_top_1000.csv', usecols=cols)
movies
```

.nlargest()
This method is equivalent to df.sort_values(columns, ascending=False).head(n), but more performant

```python
movies.nlargest(n=10, columns=['No_of_Votes'])
```

```python
movies.nlargest(n=10, columns=['IMDB_Rating','No_of_Votes'])
```

```python
movies.nsmallest(n=10, columns=['No_of_Votes'])
```

```python
movies.nsmallest(n=10, columns=['IMDB_Rating','No_of_Votes'])
```

##..keep='first' means it will keep the first duplicate(default)
##..keep='last' means it will keep the last duplicate
##..keep='all means it will keep all duplicates

```python
movies.nlargest(n=10, columns=['IMDB_Rating'], keep='all')
```
